```python
def montar_contexto_games():
    contexto = "Base histórica de jogos digitais usada pelo GameMaster IA:\n\n"

    contexto += "ERAS DOS GAMES:\n"
    for _, linha in df_eras.iterrows():
        contexto += f"- {linha['era']} ({linha['periodo']}): {linha['caracteristicas']}\n"

    contexto += "\nJOGOS DE REFERÊNCIA:\n"
    for _, linha in df_jogos.iterrows():
        contexto += f"- {linha['jogo']} | Era: {linha['era']} | Gênero: {linha['genero']}\n"

    return contexto


contexto_games = montar_contexto_games()

print(contexto_games)
```

```python
prompt = f"""
Crie uma sinopse criativa para o projeto GameMaster IA.

O GameMaster IA é uma aplicação que usa IA generativa para criar histórias,
missões e diálogos inspirados em jogos antigos, da era Atari até PS2.

Use como apoio esta base histórica:

{contexto_games}
"""

resposta = modelo.generate_content(prompt)

print(resposta.text)
```

```python
instrucao_gamemaster = """
Você é o GameMaster IA, um mestre de jogo narrativo, criativo e nostálgico.

Seu papel é criar histórias, missões, diálogos e eventos inspirados em jogos antigos,
desde a era Atari até o PlayStation 2.

Seu estilo deve misturar:
- clima de fliperama antigo;
- aventura 8-bit e 16-bit;
- mistério de jogos de PS1;
- narrativa cinematográfica de PS2.

Regras:
- Não copie personagens famosos.
- Não use nomes protegidos de franquias reais.
- Crie tudo de forma original.
- Escreva em português brasileiro.
- Seja criativo, mas mantenha organização.
"""

modelo_gamemaster = genai.GenerativeModel(
    "gemini-2.5-flash",
    system_instruction=instrucao_gamemaster
)

print("Persona GameMaster IA criada.")
```

```python
prompt_historia = f"""
Com base neste contexto histórico:

{contexto_games}

Crie uma história original de um jogo retrô chamado "Cartucho Perdido".

A história deve conter:
1. Título
2. Era de inspiração
3. Protagonista
4. Mundo do jogo
5. Vilão
6. Objetivo principal
7. Mecânicas principais
8. Final da aventura
"""

resposta_historia = modelo_gamemaster.generate_content(prompt_historia)

print(resposta_historia.text)
```

```python
instrucao_npc = """
Você é um NPC misterioso de um RPG retrô.

Você fala como um personagem antigo de jogo, com frases enigmáticas,
mas ainda compreensíveis.

Você deve parecer um personagem que vive dentro de um cartucho antigo amaldiçoado.
Suas respostas devem ser dramáticas, criativas e com clima de jogo de PS1.
"""

modelo_npc = genai.GenerativeModel(
    "gemini-2.5-flash",
    system_instruction=instrucao_npc
)

resposta_npc = modelo_npc.generate_content(
    "Um jogador acabou de entrar no mundo do cartucho perdido. Dê as boas-vindas a ele."
)

print(resposta_npc.text)
```

```python
class FichaJogo(typing.TypedDict):
    titulo: str
    era_inspiracao: str
    genero: str
    protagonista: str
    vilao: str
    objetivo: str
    mecanicas: list[str]
    publico_alvo: str
```

```python
prompt_json = """
Crie a ficha técnica de um jogo original chamado Cartucho Perdido.

O jogo deve ser inspirado em games antigos da era Atari até PS2.
Não use personagens famosos.
Crie tudo de forma original.
"""

resposta_json = modelo_gamemaster.generate_content(
    prompt_json,
    generation_config=genai.GenerationConfig(
        response_mime_type="application/json",
        response_schema=FichaJogo
    )
)

print(resposta_json.text)
```
